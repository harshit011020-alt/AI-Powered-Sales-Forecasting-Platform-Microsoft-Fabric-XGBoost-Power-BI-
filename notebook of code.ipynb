```python
train_df = spark.read.format("csv") \
    .option("header","true") \
    .option("inferSchema","true") \
    .load("Files/train.csv")

display(train_df)
```

```python
train_df.printSchema()
```

```python
# Rename columns before join
train_df = train_df.withColumnRenamed("IsHoliday", "IsHoliday_train")
walmart_df = walmart_df.withColumnRenamed("IsHoliday", "IsHoliday_features")
```

```python
from pyspark.sql.functions import to_date

# Ensure same date format
train_df = train_df.withColumn("Date", to_date("Date"))
walmart_df = walmart_df.withColumn("Date", to_date("Date"))

# Join datasets
walmart_full_df = train_df.join(
    walmart_df,
    on=["Store", "Date"],
    how="inner"
)

display(walmart_full_df)
```

```python
from pyspark.sql.functions import col

cols_to_convert = ["MarkDown1","MarkDown2","MarkDown3","MarkDown4","MarkDown5","CPI","Unemployment"]

for c in cols_to_convert:
    walmart_full_df = walmart_full_df.withColumn(c, col(c).cast("double"))
```

```python
walmart_full_df = walmart_full_df.fillna(0)
```

```python
walmart_full_df = walmart_full_df.orderBy("Store", "Date")
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import lag

window_spec = Window.partitionBy("Store").orderBy("Date")

walmart_full_df = walmart_full_df.withColumn(
    "lag_1",
    lag("Weekly_Sales", 1).over(window_spec)
)
```

```python
from pyspark.sql.functions import avg

walmart_full_df = walmart_full_df.withColumn(
    "rolling_avg",
    avg("Weekly_Sales").over(window_spec.rowsBetween(-3, 0))
)
```

```python
walmart_full_df = walmart_full_df.dropna()
```

```python
from pyspark.sql.functions import avg

walmart_full_df = walmart_full_df.withColumn(
    "rolling_avg",
    avg("Weekly_Sales").over(window_spec.rowsBetween(-3, 0))
)
```

```python
walmart_full_df = walmart_full_df.dropna()
```

```python
from pyspark.sql.functions import col

walmart_full_df = walmart_full_df.withColumn(
    "IsHoliday",
    col("IsHoliday_train")
)
```

```python
walmart_full_df = walmart_full_df.drop(
    "IsHoliday_train",
    "IsHoliday_features"
)
```

```python
walmart_final_df = walmart_full_df.select(
    "Store",
    "Dept",
    "Date",
    "Weekly_Sales",
    "lag_1",
    "rolling_avg",
    "Temperature",
    "Fuel_Price",
    "CPI",
    "Unemployment",
    "IsHoliday"
)
```

```python
walmart_final_df = walmart_final_df.dropna()
```

```python
walmart_final_df.write.mode("overwrite").saveAsTable("walmart_clean")
```

```python
spark.sql("SELECT * FROM walmart_clean LIMIT 10").show()
```

```python
walmart_df = spark.sql("SELECT * FROM walmart_clean")
display(walmart_df)
```

```python
pandas_df = walmart_df.toPandas()
```

```python
pandas_df = pandas_df.sort_values(by=["Store", "Date"])
```

```python
y = pandas_df["Weekly_Sales"]
```

```python
features = [
    "lag_1",
    "rolling_avg",
    "Temperature",
    "Fuel_Price",
    "CPI",
    "Unemployment",
    "IsHoliday"
]

X = pandas_df[features]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)
```

```python
pandas_df["Predicted_Sales"] = model.predict(X)
```

```python
spark_df = spark.createDataFrame(pandas_df)

spark_df.write.mode("overwrite").saveAsTable("walmart_predictions")
```

```python
spark.sql("SELECT * FROM walmart_predictions LIMIT 10").show()
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6
)

xgb_model.fit(X_train, y_train)
```

```python
xgb_pred = xgb_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred))
print("XGBoost RMSE:", xgb_rmse)
```

```python
rf_pred = model.predict(X_test)

rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))

print("Random Forest RMSE:", rf_rmse)
print("XGBoost RMSE:", xgb_rmse)
```

```python
final_model = xgb_model
```

```python
pandas_df["Predicted_Sales"] = final_model.predict(X)
```

```python
spark_df = spark.createDataFrame(pandas_df)

spark_df.write.mode("overwrite").saveAsTable("walmart_predictions2")
```
